```python
import pandas as pd
import os
from collections import Counter
from pprint import pprint as print
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
test=pd.read_csv('/opt/ml/teamrepo/team/output/output.csv')
result=dict(Counter(test['ans']))

myList = result.items()
myList = sorted(myList) 
x, y = zip(*myList) 

plt.plot(x, y)
plt.show()
```
